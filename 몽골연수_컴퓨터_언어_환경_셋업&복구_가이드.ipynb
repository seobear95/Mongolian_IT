재능대 컴퓨터소프트웨어학과 서연경 교수


Chrome이나 Office 등의 언어 설정을 하기 전에 시스템 언어를 몽골어로 변경하면, 메뉴를 찾기 어려울 수 있습니다. 따라서, 크롬과 오피스 등의 언어 설정을 먼저 변경하고 마지막에 시스템 언어를 몽골어로 변경하는 것이 좋습니다. 이렇게 하면 필요한 모든 설정을 이해할 수 있는 언어로 변경한 후, 시스템 언어를 변경하여 불편함을 최소화할 수 있습니다.

### 몽골어 환경 설정 가이드

#### 1. 몽골어 언어팩 및 키보드 설치

**Step 1: Windows 설정 열기**
1. **시작 버튼**을 클릭합니다.
2. **설정** 아이콘을 클릭하여 설정 창을 엽니다.

**Step 2: 언어 설정**
1. **시간 및 언어**를 클릭합니다.
2. 왼쪽 메뉴에서 **언어**를 선택합니다.
3. **언어 추가** 버튼을 클릭합니다.

**Step 3: 몽골어 추가**
1. 검색창에 `Mongolian`을 입력하고 목록에서 몽골어를 선택합니다.
2. **다음**을 클릭합니다.
3. **언어 기능 설치** 화면에서 필요한 옵션을 선택하고 **설치**를 클릭합니다.

**Step 4: 몽골어 키보드 추가**
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 언어 목록에서 추가한 몽골어를 선택하고 **옵션** 버튼을 클릭합니다.
3. **키보드 추가** 버튼을 클릭하고 `Mongolian`을 선택하여 추가합니다.

**Step 5: 언어 및 키보드 전환 설정**
1. **설정** > **장치** > **입력**으로 이동합니다.
2. 오른쪽에서 **고급 키보드 설정**을 클릭합니다.
3. **입력 언어 전환 방법 변경**을 클릭합니다.
4. **입력 언어 전환 단축키 변경**에서 단축키를 설정합니다. (기본값은 `Alt` + `Shift`입니다.)

### 소프트웨어 설정

#### Microsoft Office 설정
**Step 1: Office 언어 설정**
1. Microsoft Word 또는 다른 Office 프로그램을 엽니다.
2. **파일** 메뉴를 클릭하고 **옵션**을 선택합니다.
3. **언어** 탭을 클릭합니다.
4. **Office 편집 언어 추가** 버튼을 클릭하고 `몽골어`를 선택합니다.
5. **표시 및 도움말 언어** 섹션에서 `몽골어`를 선택합니다.
6. **기본값으로 설정** 버튼을 클릭하고 **확인**을 클릭합니다.
7. Office 프로그램을 재시작합니다.

#### Google Chrome 설정
**Step 1: 언어 설정**
1. Google Chrome을 엽니다.
2. 오른쪽 상단의 **더보기** 아이콘(점 세 개)을 클릭하고 **설정**을 선택합니다.
3. **고급**을 클릭하여 고급 설정을 엽니다.
4. **언어** 섹션을 찾습니다.
5. **언어 추가** 버튼을 클릭하고 `Mongolian`을 선택하여 추가합니다.
6. 추가한 `몽골어` 옆의 **더보기** 아이콘을 클릭하고 **Google Chrome 표시 언어로 설정**을 선택합니다.
7. Chrome을 재시작합니다.

#### Adobe Acrobat Reader 설정
**Step 1: Adobe Acrobat Reader 열기**
1. Adobe Acrobat Reader를 엽니다.
2. **편집** 메뉴를 클릭하고 **기본 설정**을 선택합니다.
3. 왼쪽 메뉴에서 **언어**를 선택합니다.
4. **애플리케이션 언어**를 `시스템 기본 언어 사용`으로 설정하거나 `몽골어`로 직접 설정합니다.
5. **확인**을 클릭하고 프로그램을 재시작합니다.

### 6. 시스템 표시 언어 변경

**Step 1: 시스템 표시 언어 변경**
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 언어 목록에서 `몽골어`를 클릭한 다음 **기본값으로 설정** 버튼을 클릭합니다.
3. **설정** > **시간 및 언어** > **지역 및 언어**로 이동하여 `Windows 표시 언어`를 `몽골어`로 설정합니다.
4. 시스템을 재부팅하여 변경 사항을 적용합니다.

몽골 학생들의 연수가 끝난 후, 한국어 및 한국 학생들이 다시 수업을 원활하게 진행할 수 있도록 노트북을 원래 상태로 복원하는 절차를 자세히 안내드리겠습니다.
시스템 언어를 변경해도 Office 및 기타 프로그램의 언어 설정이 자동으로 변경되지 않기 때문에, 개별적으로 각 프로그램의 언어 설정을 변경해야 합니다. 따라서, 재부팅 후에도 Office 툴들은 설정을 변경하지 않으면 계속 몽골어로 표시될 수 있습니다. 아래는 재부팅 후 Office 및 기타 소프트웨어의 언어 설정을 변경하는 방법을 포함한 전체 가이드입니다. 예) 몽골어(한국어) 메뉴 흐름 참고

### 몽골어 환경에서 원래 상태로 복원하는 가이드 (한국어로 변경)

#### 1. Windows 표시 언어 변경 (몽골어 메뉴에서 한국어로 변경)

**Step 1: Windows 설정 열기 (Тохиргоо нээх)**
1. **Эхлэх** (시작) 버튼을 클릭합니다.
2. **Тохиргоо** (설정) 아이콘을 클릭하여 설정 창을 엽니다.

**Step 2: 언어 설정 (Хэл тохиргоо)**
1. **Цаг хугацаа ба хэл** (시간 및 언어)를 클릭합니다.
2. 왼쪽 메뉴에서 **Хэл** (언어)를 선택합니다.
3. **Суурилуулах хэл нэмэх** (언어 추가) 버튼을 클릭합니다.

**Step 3: 한국어 추가 (Солонгос хэл нэмэх)**
1. 검색창에 `Korean`을 입력하고 목록에서 `Солонгос хэл` (한국어)를 선택합니다.
2. **Дараах** (다음)을 클릭합니다.
3. **Хэлний багцыг суурилуулах** (언어 기능 설치) 화면에서 필요한 옵션을 선택하고 **Суурилуулах** (설치)를 클릭합니다.

**Step 4: 시스템 표시 언어 변경 (Системийн хэл солих)**
1. **Цаг хугацаа ба хэл** (시간 및 언어) > **Хэл** (언어)로 이동합니다.
2. 언어 목록에서 **Солонгос хэл** (한국어)를 클릭한 다음 **Үндсэн болгон тохируулах** (기본값으로 설정) 버튼을 클릭합니다.
3. **Цаг хугацаа ба хэл** (시간 및 언어) > **Бүс ба хэл** (지역 및 언어)로 이동하여 **Windows-ийн дэлгэцийн хэл** (Windows 표시 언어)를 **Солонгос хэл** (한국어)로 설정합니다.
4. 시스템을 재부팅하여 변경 사항을 적용합니다.

### 소프트웨어 언어 설정 변경 (한국어로 복원)

#### Microsoft Office 설정

**Step 1: Microsoft Office 열기 (Microsoft Office нээх)**
1. **Microsoft Word** 또는 다른 Office 프로그램을 엽니다.

**Step 2: 파일 메뉴 (Файл цэс)**
1. **Файл** (파일) 메뉴를 클릭합니다.
2. **Сонголт** (옵션)를 선택합니다.

**Step 3: 언어 설정 (Хэл тохиргоо)**
1. **Хэл** (언어) 탭을 클릭합니다.
2. **Office-ийн засварын хэл нэмэх** (Office 편집 언어 추가) 버튼을 클릭하고 `한국어`를 선택합니다.
3. **Дэлгэц ба тусламжийн хэл** (표시 및 도움말 언어) 섹션에서 `한국어`를 선택합니다.
4. **Үндсэн болгон тохируулах** (기본값으로 설정) 버튼을 클릭합니다.
5. **Батлах** (확인)을 클릭하고 Office 프로그램을 재시작합니다.

#### Google Chrome 설정

**Step 1: Google Chrome 열기 (Google Chrome нээх)**
1. **Google Chrome**을 엽니다.

**Step 2: 설정 메뉴 (Тохиргоо цэс)**
1. 오른쪽 상단의 **Цэс** (더보기) 아이콘(점 세 개)을 클릭하고 **Тохиргоо** (설정)를 선택합니다.

**Step 3: 언어 설정 (Хэл тохиргоо)**
1. **Нарийвчилсан** (고급)을 클릭하여 고급 설정을 엽니다.
2. **Хэл** (언어) 섹션을 찾습니다.
3. **Хэл нэмэх** (언어 추가) 버튼을 클릭하고 `Korean`을 선택하여 추가합니다.
4. 추가한 `Солонгос хэл` (한국어) 옆의 **Дэлгэрэнгүй** (더보기) 아이콘을 클릭하고 **Google Chrome-ийн дэлгэцийн хэл болгон тохируулах** (Google Chrome 표시 언어로 설정)를 선택합니다.
5. Chrome을 재시작합니다.

#### Adobe Acrobat Reader 설정

**Step 1: Adobe Acrobat Reader 열기 (Adobe Acrobat Reader нээх)**
1. **Adobe Acrobat Reader**를 엽니다.

**Step 2: 편집 메뉴 (Засах цэс)**
1. **Засах** (편집) 메뉴를 클릭하고 **Тохиргоо** (기본 설정)를 선택합니다.

**Step 3: 언어 설정 (Хэл тохиргоо)**
1. 왼쪽 메뉴에서 **Хэл** (언어)를 선택합니다.
2. **Хэрэглээний хэл** (애플리케이션 언어)를 `Системийн үндсэн хэл ашиглах` (시스템 기본 언어 사용)으로 설정하거나 `Солонгос хэл` (한국어)로 직접 설정합니다.
3. **Батлах** (확인)을 클릭하고 프로그램을 재시작합니다.

이 가이드를 따라 설정을 변경하고 복원하면, 몽골 학생들이 연수를 마친 후 한국 학생들이 원활하게 수업을 진행할 수 있습니다. 추가적인 질문이나 도움이 필요하면 언제든지 문의해 주세요.
