재능대 컴퓨터소프트웨어학과 서연경 교수


### Windows에서 몽골어 자판과 언어팩 추가를 위한 단계별 가이드

몽골 학생들이 Windows 노트북에서 몽골어 자판과 언어팩을 추가할 수 있도록 단계별로 안내하겠습니다.

### 1. 윈도우 언어팩 설치

#### Step 1: Windows 설정 열기
1. **시작 버튼**을 클릭합니다.
2. **설정** 아이콘을 클릭하여 설정 창을 엽니다.

#### Step 2: 언어 설정
1. **시간 및 언어**를 클릭합니다.
2. 왼쪽 메뉴에서 **언어**를 선택합니다.
3. **언어 추가** 버튼을 클릭합니다.

#### Step 3: 몽골어 추가
1. 검색창에 `Mongolian`을 입력하고 목록에서 몽골어를 선택합니다.
2. **다음**을 클릭합니다.
3. **언어 기능 설치** 화면에서 필요한 옵션을 선택하고 **설치**를 클릭합니다.

### 2. 몽골어 키보드 추가

#### Step 1: 언어 옵션 열기
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 언어 목록에서 추가한 몽골어를 선택하고 **옵션** 버튼을 클릭합니다.

#### Step 2: 몽골어 키보드 추가
1. **키보드 추가** 버튼을 클릭합니다.
2. **Mongolian**을 선택하여 추가합니다.

### 3. 언어 및 키보드 전환 설정

#### Step 1: 언어 전환 단축키 설정
1. **설정** > **장치** > **입력**으로 이동합니다.
2. 오른쪽에서 **고급 키보드 설정**을 클릭합니다.
3. **입력 언어 전환 방법 변경**을 클릭합니다.
4. **입력 언어 전환 단축키 변경**에서 단축키를 설정합니다. 기본값은 `Alt` + `Shift`입니다.

### 4. 시스템 표시 언어 변경

#### Step 1: 기본 언어 설정
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 몽골어를 클릭한 다음 **기본값으로 설정** 버튼을 클릭합니다.

### 5. 소프트웨어 설정 (Microsoft Office 예시)

#### Step 1: Office 언어 설정
1. Microsoft Word 또는 다른 Office 프로그램을 엽니다.
2. **파일** 메뉴를 클릭하고 **옵션**을 선택합니다.
3. **언어** 탭을 클릭합니다.
4. **Office 편집 언어 추가** 버튼을 클릭하고 몽골어를 선택합니다.

#### Step 2: 표시 언어 설정
1. **표시 및 도움말 언어** 섹션에서 몽골어를 선택합니다.
2. **기본값으로 설정** 버튼을 클릭합니다.
3. **확인**을 클릭하고 Office 프로그램을 재시작합니다.

### 6. 웹 브라우저 언어 설정 (Google Chrome 예시)

#### Step 1: 언어 설정
1. Google Chrome을 엽니다.
2. 오른쪽 상단의 **더보기** 아이콘(점 세 개)을 클릭하고 **설정**을 선택합니다.
3. **고급**을 클릭하여 고급 설정을 엽니다.
4. **언어** 섹션을 찾습니다.

#### Step 2: 몽골어 추가
1. **언어 추가** 버튼을 클릭합니다.
2. **Mongolian**을 선택하고 **추가**를 클릭합니다.
3. 추가한 몽골어 옆의 **더보기** 아이콘을 클릭하고 **Google Chrome 표시 언어로 설정**을 선택합니다.
4. Chrome을 재시작합니다.

### 7. PDF 리더 언어 설정 (Adobe Acrobat Reader 예시)

#### Step 1: Adobe Acrobat Reader 열기
1. Adobe Acrobat Reader를 엽니다.
2. **편집** 메뉴를 클릭하고 **기본 설정**을 선택합니다.

#### Step 2: 언어 설정 변경
1. 왼쪽 메뉴에서 **언어**를 선택합니다.
2. **애플리케이션 언어**를 **시스템 기본 언어 사용**으로 설정하거나 **몽골어**로 직접 설정합니다.
3. **확인**을 클릭하고 프로그램을 재시작합니다.

### 마무리

--------------------------------------------

몽골 학생들의 연수가 끝난 후, 한국어 및 한국 학생들이 다시 수업을 원활하게 진행할 수 있도록 노트북을 원래 상태로 복원하는 절차를 자세히 안내드리겠습니다.

### 원복 절차

#### 1. Windows 표시 언어 변경

##### Step 1: 설정 열기
1. **시작 버튼**을 클릭합니다.
2. **설정** 아이콘을 클릭하여 설정 창을 엽니다.

##### Step 2: 언어 설정
1. **시간 및 언어**를 클릭합니다.
2. 왼쪽 메뉴에서 **언어**를 선택합니다.
3. **기본 언어** 섹션에서 `한국어`를 선택합니다.
4. **기본값으로 설정** 버튼을 클릭합니다.

##### Step 3: 표시 언어 변경
1. **설정** > **시간 및 언어** > **지역 및 언어**로 이동합니다.
2. `Windows 표시 언어`를 `한국어`로 설정합니다.
3. 시스템을 재부팅하여 변경 사항을 적용합니다.

#### 2. 한글 자판 추가

##### Step 1: 언어 옵션 열기
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 언어 목록에서 `한국어`를 선택하고 **옵션** 버튼을 클릭합니다.

##### Step 2: 한글 키보드 추가
1. **키보드 추가** 버튼을 클릭합니다.
2. **Microsoft IME**를 선택하여 추가합니다.

#### 3. 불필요한 언어 제거

##### Step 1: 언어 목록에서 제거
1. **설정** > **시간 및 언어** > **언어**로 이동합니다.
2. 언어 목록에서 `몽골어`를 선택하고 **제거** 버튼을 클릭합니다.

#### 4. 소프트웨어 언어 설정 변경 (Microsoft Office 예시)

##### Step 1: Office 언어 설정
1. Microsoft Word 또는 다른 Office 프로그램을 엽니다.
2. **파일** 메뉴를 클릭하고 **옵션**을 선택합니다.
3. **언어** 탭을 클릭합니다.

##### Step 2: 표시 언어 변경
1. **Office 편집 언어 추가** 버튼을 클릭하고 `한국어`를 선택합니다.
2. **표시 및 도움말 언어** 섹션에서 `한국어`를 선택합니다.
3. **기본값으로 설정** 버튼을 클릭합니다.
4. **확인**을 클릭하고 Office 프로그램을 재시작합니다.

#### 5. 웹 브라우저 언어 설정 변경 (Google Chrome 예시)

##### Step 1: 언어 설정
1. Google Chrome을 엽니다.
2. 오른쪽 상단의 **더보기** 아이콘(점 세 개)을 클릭하고 **설정**을 선택합니다.
3. **고급**을 클릭하여 고급 설정을 엽니다.
4. **언어** 섹션을 찾습니다.

##### Step 2: 한국어 추가
1. **언어 추가** 버튼을 클릭합니다.
2. **Korean**을 선택하고 **추가**를 클릭합니다.
3. 추가한 한국어 옆의 **더보기** 아이콘을 클릭하고 **Google Chrome 표시 언어로 설정**을 선택합니다.
4. Chrome을 재시작합니다.

#### 6. PDF 리더 언어 설정 변경 (Adobe Acrobat Reader 예시)

##### Step 1: Adobe Acrobat Reader 열기
1. Adobe Acrobat Reader를 엽니다.
2. **편집** 메뉴를 클릭하고 **기본 설정**을 선택합니다.

##### Step 2: 언어 설정 변경
1. 왼쪽 메뉴에서 **언어**를 선택합니다.
2. **애플리케이션 언어**를 **시스템 기본 언어 사용**으로 설정하거나 **한국어**로 직접 설정합니다.
3. **확인**을 클릭하고 프로그램을 재시작합니다.

### 네트워크 및 기타 설정 복원

#### Step 1: 네트워크 설정 확인
1. **설정** > **네트워크 및 인터넷**으로 이동합니다.
2. Wi-Fi 및 LAN 설정을 확인하여 한국 내 네트워크에 제대로 연결되었는지 확인합니다.

#### Step 2: 기타 프로그램 설정
1. 각 프로그램의 언어 및 지역 설정을 확인하여 필요한 경우 한국어로 변경합니다.
2. 필요한 경우 프로그램을 재설치하여 초기 상태로 복원합니다.

위의 절차를 따라 설정을 복원하면, 한국 학생들이 노트북을 원활하게 사용할 수 있습니다. 이 가이드를 인쇄하거나 디지털 파일로 제공하여 담당자가 따라할 수 있도록 하면 좋습니다. 추가적인 도움이 필요하면 언제든지 문의해 주세요.
